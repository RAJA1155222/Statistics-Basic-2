✅ Q2. Difference between mean, median and mode
Measure	Meaning	Use
Mean	Average of all values	Used when data has no extreme values
Median	Middle value after sorting	Used when data has outliers
Mode	Most frequently occurring value	Used to find the most common value

✅ Q5. How measures of dispersion are used

Measures of dispersion show how spread the data is.

Range → max − min

Variance → average of squared deviations from mean

Standard deviation → square root of variance

Example
If two classes have the same average marks but one class has higher standard deviation, then that class has more variation in marks.

✅ Q8. What is skewness?

Skewness describes the asymmetry of a data distribution.

Left skewed → tail on the left

Right skewed → tail on the right

✅ Q9. In right-skewed data, position of median with respect to mean

For right skewed data:

Mean > Median

✅ Q10. Difference between covariance and correlation
Covariance	Correlation
Shows direction of relationship	Shows direction and strength
No fixed scale	Value lies between −1 and +1
Hard to interpret	Easy to interpret

Both are used to study how two variables change together.

✅ Q12. Relationship between measures of central tendency in normal distribution

For a normal distribution:

Mean = Median = Mode

✅ Q13. How is covariance different from correlation?

Covariance only tells the direction of relationship.

Correlation tells both direction and strength of relationship and is standardized.

✅ Q14. Effect of outliers on measures of central tendency and dispersion

Outliers:

Strongly affect mean

Strongly increase standard deviation and variance

Have little effect on median

Example

Data: [10, 12, 11, 13, 100]

Mean increases due to 100

Median remains close to original values
